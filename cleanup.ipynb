```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
import googlemaps

#This won't work without your google geocode API Key
gmaps = googlemaps.Client(key='')


def getCounty(location):
    county = np.NaN
    geocode_result = gmaps.geocode(location)
    if len(geocode_result)>0:
        counties = filter(lambda x: "administrative_area_level_2" in x["types"],geocode_result[0]["address_components"])
        if len(counties)>0:
            county = counties[0]["long_name"]
    return county



def getCity(location):
    city = np.NaN
    geocode_result = gmaps.geocode(location)
    if len(geocode_result)>0:
        cities = filter(lambda x: "locality" in x["types"],geocode_result[0]["address_components"])
        if len(cities)>0:
            city = cities[0]["long_name"]
    return city
```

```python
d1 = pd.read_csv("Data\CSV\Amazon-Delivery-Station-Network-in-USA.csv")
d2 = pd.read_csv("Data\CSV\Amazon-Food-Distribution-Center-Network-in-USA.csv")
d3 = pd.read_csv("Data\CSV\Amazon-Fulfillment-Center-Network-in-USA.csv")
d4 = pd.read_csv("Data\CSV\Amazon-Inbound-Crossdock-Network-in-USA.csv")
d5 = pd.read_csv("Data\CSV\Amazon-Prime-Now-Network-in-USA.csv")
d6 = pd.read_csv("Data\CSV\Amazon-Sortation-Center-Network-in-USA.csv")

```

```python
display(d1.head())
display(d2.head())
display(d3.head())
display(d4.head())
display(d5.head())
display(d6.head())
```

```python
#add a column for type to each one 
#combine into one oframe
d1["Type"]="Delivery Station"
d2["Type"]="Fresh"
d3["Type"]="Fulfillment"
d4["Type"]="Crossdock"
d5["Type"]="Now"
d6["Type"]="Sortation"
```

```python
d_concat = pd.concat([d1,d2,d3,d4,d5,d6])
```

```python
print(d_concat.shape)
print (d_concat.shape)
d_concat.head()
```

```python
#get rid of weird new lines in column names
d_reset = d_concat.reset_index(drop=True)

good_cols = d_reset.columns.str.replace("\n"," ")
d_reset.columns=good_cols
```

```python
d_reset.head()
```

```python
d_reset.columns
```

```python
#replace the \ns in year and then figure out how large a date range we need
#do the same thing for Location 
d_reset["Year Opened"]=d_reset["Year Opened"].str.replace("\n", " ")
d_reset["Location"]=d_reset["Location"].str.replace("\n", " ")
```

```python
#some of them are missing years, find out which 
d_reset[d_reset["Year Opened"].str.contains("\d{4}").fillna(False)==False]
```

```python
d_reset['County'] = np.vectorize(getCounty,otypes=[object])(d_reset["Location"])
d_reset['City'] = np.vectorize(getCity,otypes=[object])(d_reset["Location"])
```

```python
#split out the year to new columns
d_reset["Year"] = d_reset["Year Opened"].str.extract("(\d{4})",expand=False).astype(int,errors="ignore")
```

```python
years = d_reset["Year Opened"].str.extract("(\d{4})",expand=False).sort_values()
```

```python
#some of them are future openings
d_reset[d_reset["Year Opened"].str.contains("2019").fillna(False)]
```

```python
#save output
d_reset.to_csv("amazon_labelled_locations.csv",index=False)
```
